```python
import pandas as pd
```

```python
# Reading the xlsm file as a dataframe in pandas
world_pop = pd.read_excel('world-population.xlsm', sheet_name='world-population')
```

```python
type(world_pop)
```

```python
world_pop
```

```python
import matplotlib.pyplot as plt

year = world_pop['Year']
pop = world_pop['Population']/1000000000

plt.plot(year,pop)
plt.title('Line chart of world population growth by year')
plt.xlabel('Year')
plt.ylabel('Population (in billions)')
plt.show()
```
