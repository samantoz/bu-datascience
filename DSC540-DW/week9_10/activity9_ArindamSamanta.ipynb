```python
import urllib.request, urllib.parse, urllib.error
import requests
from bs4 import BeautifulSoup as soup
import ssl
import re
```

```python
# Checking the SSL certificate
# Ignore ssl certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
# Read the HTML from the URL and pass on to BeautifulSoup
top100url = 'https://www.gutenberg.org/browse/scores/top'
response = requests.get(top100url)
```

```python
# check the status of the web request
def status_check(r):
    if r.status_code==200:
        print("Success!")
        return 1
    else:
        print("Failed!")
        return -1
    
```

```python
# check the status of the response
status_check(response)
```

```python
contents = response.content.decode(response.encoding)
page_contents = soup(contents, "html.parser")
page_contents
```

```python
# empty list to hold all the http links in the HTML page
lst_links=[]
# find the href tags and store them in a list of links
for link in page_contents.findAll('a'):
    lst_links.append(link.get('href'))
```

```python
lst_links[:40]
```

```python
# use a regular expression to find the numeric digits in these links.
# numbers 19 to 118 in the original list of links have the top 100 eBooks.
booknum=[]
for i in range(19,119):
    link=lst_links[i].strip()
    # regular expression to find the numeric digits in the link
    n=re.findall('[0-9]+',link)
    if len(n)==1:
        #append the filenumber 
        booknum.append(int(n[0]))
```

```python
#print the file numbers
print("\nThe file numbers for the top 100 ebooks on Gutenberg are shown below\n"+"-"*70)
print(booknum)
```

```python
#search the extracted text from the soup object to find the names of the top 100 eBooks.
# Temp empty list of Ebooks names
lst_titles_temp=[]
# use splitlines method of soup.text to split the lines of the text
# start_idx=page_contents.text.splitlines().index('Top 100 EBooks yesterday')
start_idx = [index for index, value in enumerate(page_contents.text.splitlines()) if value =='Top 100 EBooks yesterday']
# loop 1-100 to add the strings to the list
for i in range(100):
    lst_titles_temp.append(page_contents.text.splitlines()[start_idx[1]+2+i])
```

```python
# use a regular expression to extract only the text from the name strings.
# using match and span to find the indices
lst_titles=[]
for i in range(100):
    id1,id2=re.match('^[a-zA-Z ]*',lst_titles_temp[i]).span()
    lst_titles.append(lst_titles_temp[i][id1:id2])
```

```python
#print the list of titles
for l in lst_titles:
    print(l)
```
